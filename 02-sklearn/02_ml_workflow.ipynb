```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

# Load the data set
iris = load_iris()

# iris.data = features(ndarray) of the data set
iris_data = iris.data

# iris.target = labels(ndarray) of the data set
# iris.target_names = names of labels
iris_label = iris.target
print('iris target값:', iris_label)
print('iris target명:', iris.target_names)

# iris.featrue_names = names of features
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)
```

```python
# iris.featrue_names = names of features
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)
```

## 1. 데이터 셋 분리

```python
# test_size = 0.2 : 테스트 셋 20%, 학습 셋 80%
# random_state : 호출할 때마다 같은 학습/테스트 데이터 셋을 생성하기 위한 난수 값
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size = 0.2, random_state = 11)
```

## 2. 학습

```python
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state = 11)
```

```python
# 학습 수행
dt_clf.fit(X_train, y_train)
```

## 3. 테스트

```python
# 테스트 수행
pred = dt_clf.predict(X_test)
```

## 4. 성능 평가

```python
from sklearn.metrics import accuracy_score
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test, pred)))
```
